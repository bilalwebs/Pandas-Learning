```python
import pandas as pd

delivery = pd.read_csv('deliveries.csv')
```

```python

delivery.head()
```

```python
runs = delivery.groupby('batsman')
```

```python
runs
```

```python
runs['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
cricket_team = delivery.groupby('batsman')
cricket_team.get_group('V Kohli')
```

```python
# Step 1: Grouping
cricket_team = delivery.groupby('batsman')

# Step 2: Kohli ka data alag kiya
kohli_data = cricket_team.get_group('V Kohli')

# Step 3: Target column par sum lagaya
total_runs = kohli_data['batsman_runs'].sum()

print(total_runs)
```

```python
wide_run = delivery[delivery['wide_runs'] > 0]
wide_run.groupby('bowler')['wide_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# 1. Pehle mask banayein (Row filter)
mask = delivery['wide_runs'] > 0

# 2. Mask ke baad column ki list dein
specific_data = delivery[mask][['bowler', 'wide_runs']]

# Ab check karein
print(specific_data.head())
```

```python
mera_mask = delivery['wide_runs']>0
print(mera_mask.head())

filtered_data = delivery[mera_mask][['bowler', 'wide_runs']]
filtered_data.head()
```

```python
mask = delivery[delivery['wide_runs']>0]
group_run=mask.groupby('bowler')['wide_runs'].sum().sort_values(ascending=False).head()
print(group_run)
```

```python
dhoni_data = delivery.groupby('batsman')
dhoni_data.get_group('MS Dhoni')
```

```python
all_batsman = delivery.groupby('batsman')

kholi = all_batsman.get_group('V Kohli')['batsman_runs'].sum()
dhoni = all_batsman.get_group("MS Dhoni")['batsman_runs'].sum()

print(f"Kohli: {kholi}, Dhoni: {dhoni}")
```

```python
# 1. Mask lagaya (Sirf in do players ka data)
mask = delivery['batsman'].isin(['V Kohli', 'MS Dhoni'])
dual_data = delivery[mask]

# 2. Groupby kar ke sum kiya
comparison = dual_data.groupby('batsman')['batsman_runs'].sum()

print(comparison)
```

```python
over = delivery[delivery['over']<=6]

batting_team = over.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
print(batting_team)
```

```python
death_overs_data = delivery[delivery['over']>16]

top_death_batsmen = death_overs_data.groupby('batsman')['total_runs'].sum().sort_values(ascending=False).head()
print(top_death_batsmen)
```

```python
player_dismissed = delivery[delivery['player_dismissed'].notnull()]
bowler = player_dismissed.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head()

bowler
```

```python
player_dismissed = delivery[delivery['player_dismissed'] != 'nan']
bowler = player_dismissed.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head()

bowler
```

```python
# Sirf woh dismissals jo bowler ki profile mein jati hain
vaild_dismissals = ['caught', 'bowled', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']

# Filter data
wicket_data = delivery[delivery['dismissal_kind'].isin(vaild_dismissals)]

# Ab count karein
real_wickets = wicket_data.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False)
```
