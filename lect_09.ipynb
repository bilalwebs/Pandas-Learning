```python
# The sort_values() method
```

```python
import pandas as pd

data = pd.read_csv("matches.csv")
data.head()
```

```python
(data["team2"].value_counts() + data["team1"].value_counts()).head()
```

```python
(data["team2"].value_counts() + data["team1"].value_counts()).sort_values(ascending=False).head()
```

```python
data.sort_values("city")
```

```python
data.sort_values("city",ascending=False) # Sort in descending order (large to Small)
```

```python
data.sort_values(by=['city','date'],ascending=[True,False]) # Sort by city and then by date
```

```python
# sort_index() method is used to sort the DataFrame based on its index labels. It sorts the DataFrame in ascending order by default, but you can specify descending order as well.
```

```python
# Step 1: Wins count karein (Yeh counts ke hisab se sorted hoga)
winners = data['winner'].value_counts()

# Step 2: Teams ko alphabetical order mein laayein
sorted_teams = winners.sort_index()

print(sorted_teams)
```

```python
# Z se A ki taraf teams ko sort karein
data['winner'].value_counts().sort_index(ascending=False)
```

```python
import pandas as pd

# Dummy Data: Teams aur unki total wins
wins = pd.Series([50, 85, 60], index=['Kolkata', 'Bangalore', 'Delhi'])
print("Asli Data:")
print(wins)
```

```python
# Numbers ko bade se chota (85, 60, 50) karein
result_values = wins.sort_values(ascending=False)
print(result_values)
```

```python
# Index (B, D, K) ko tartib mein laayein
result_index = wins.sort_index()
print(result_index)
```

```python
# sortvalues() method se values ko sort karte hain, jabki sort_index() method se index labels ko sort karte hain.

# syntx: sort_values(ascending=True/False, inplace=True/False)
```
