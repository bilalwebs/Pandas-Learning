```python
import pandas as pd
delivery = pd.read_csv("deliveries.csv")
delivery.head()
```

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
# all_batsman[all_batsman>200].index
# all_batsman[all_batsman>200].values
# all_batsman[all_batsman>200].values.tolist()
# all_batsman[all_batsman>200].index.tolist()
x=all_batsman[all_batsman>200]
batsman_list = x.index.tolist()
batsman_list
```

```python
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
final
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
runs
```

```python
balls = final.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
balls
```

```python
sr = (runs/balls)*100
sr
```

```python
target_players = ['CH Gayle', 'MS Dhoni', 'RG Sharma']

mask = delivery['batsman'].isin(target_players)
top_players_data = delivery[mask]

print("Selected Players Data:")
top_players_data.groupby('batsman')['batsman_runs'].sum()
```

```python
# 4. NOT ISIN (~): Maan lijiye humein in 3 ke ILAWA baaki sab ka data chahiye
other_players_mask = ~delivery['batsman'].isin(target_players)
others_data = delivery[other_players_mask]

print("\nBaaki saare players ka count:")
print(others_data['batsman'].nunique())
```
