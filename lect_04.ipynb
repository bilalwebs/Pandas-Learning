````python
# ## iloc (Integer Location)

# `iloc` is used to fetch rows and columns by their position (index) in a DataFrame.  
# It takes two parameters:

# - First → Rows
# - Second → Columns  

# Both parameters can take:
# - A single value
# - A list of values
# - A range (slice)

# ---

# ## Definition

# The `iloc` indexer is used to select data from a DataFrame using integer-based positions.  
# It works like giving exact coordinates (row number and column number).

# ---

# ## Key Rules of iloc

# ### 1. Numbers Only
# - Only accepts integers (0, 1, 2, 3, ...)
# - Does NOT accept column names like `"team1"` or `"city"`

# ---

# ### 2. 0-Based Indexing
# - Index starts from 0
# - First row → 0  
# - First column → 0  

# ---

# ### 3. Exclusion Rule
# - When using a range like `0:5`
# - It includes: 0, 1, 2, 3, 4  
# - It excludes: 5

# ---

# ## Example Usage

# ```python
# df.iloc[0, 1]        # First row, second column

# df.iloc[0:3, 0:2]   # Rows 0–2 and columns 0–1

# df.iloc[[0,2], [1,3]]  # Specific rows and columns
````

````python
data.iloc[0] # Fetch the first row of the dataframe
````

````python
data.iloc[1:3]
````

````python
data.iloc[1:11:2] # Fetch every second row from the second to the tenth row of the dataframe
````

````python
# Fancy Indexing 

data.iloc[[1,5,9]] # Fetch the second, sixth and tenth row of the dataframe
````

````python
data.iloc[:,[1,2,3]] # Fetch the columns team1, team2 and winner of the dataframe
````
