```python
# # The Groupby() function

# Definition of GroupBy
# In Pandas, the groupby() function is used to split the data into groups based on some criteria. It allows you to group rows that have the same values in a particular column and then perform some mathematical operation (like sum, average, or count) on another column.

# The Main Concept: "Split-Apply-Combine"
# Yeh groupby ka sab se barha concept hai. Iske 3 main steps hotay hain:

# Split:
# Breaking the dataset into smaller groups based on a specific column.
# (Data ko tukron mein baantna kisi ek column ki wajah se).

# Apply:
# Computing a mathematical function (like sum(), mean(), count()) on each group independently.
# (Har tukray par alag se calculation karna).

# Combine:
# Merging the results back into a single data structure (like a new DataFrame or Series).
# (Saray results ko wapis jod kar ek table bana dena).

# Syntax Breakdown
# Python
# df.groupby('Column_to_Group_By')['Target_Column'].function()
# df: Aapka dataset (DataFrame).

# groupby('Column_to_Group_By'): Woh column jis mein categories hain (e.g., City, Gender, Sector).

# ['Target_Column']: Woh column jis par aapko calculation karni hai (e.g., Sales, Profit, Age).

# .function(): Woh calculation jo aap karna chahte hain (e.g., .sum(), .mean(), .max()).
```

```python
import pandas as pd

# Data load karein
data = pd.read_csv("Fortune_500.csv", encoding='latin1')

def clean_currency(value):
    if isinstance(value, str):
        # 1. $, commas aur faltu spaces khatam karein
        clean_val = value.replace('$', '').replace(',', '').strip()
        
        # 2. Agar value bracket mein hai (jaise loss), to minus sign lagayein
        if clean_val.startswith('(') and clean_val.endswith(')'):
            clean_val = '-' + clean_val[1:-1]
        
        # 3. Agar value sirf '-' hai ya khali hai, to usay 0.0 kardein
        if clean_val == '-' or clean_val == '':
            return 0.0
            
        try:
            return float(clean_val)
        except ValueError:
            # Agar koi aur ajeeb cheez hai jo number nahi ban sakti
            return 0.0
    return value

# Cleaning apply karein
data['Revenues'] = data['Revenues'].apply(clean_currency)
data['Profits'] = data['Profits'].apply(clean_currency)

print("Data Clean ho gaya hai! Ab 'Revenues' aur 'Profits' numbers ban chuke hain.")
```
