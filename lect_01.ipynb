## Pandas

Pandas is a Python library used for data manipulation and analysis.  
It helps in working with structured data like tables and time series.

---

## Pandas Main Data Structures

### 1. DataFrame
A DataFrame is a 2-dimensional data structure (like a table).

- It has rows and columns
- Similar to an Excel sheet or SQL table
- Can store different types of data (numbers, strings, etc.)

Example:

| Name | Age |
|------|-----|
| Ali  | 20  |
| Sara | 22  |

---

### 2. Series
A Series is a 1-dimensional data structure.

- It is like a single column
- Can hold a list of values

Example:
Age:
0 20
1 22
2 25

---

## Summary

- Series → Single column (1D)
- DataFrame → Full table (2D)
