```python
# The Plot function
```

```python
import pandas as pd
```

```python
data = pd.read_csv("matches.csv")
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.show()
```

```python
data['city'].value_counts().head(5).plot(kind='pie')
```

```python
# Top 5 Teams with most wins
data['winner'].value_counts().head(5).plot(kind='bar', color='skyblue')
```

```python
# Toss decisions ratio
data['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
# Number of matches played each season
data['season'].value_counts().sort_index().plot(kind="line",marker='o',color="green")
```

```python
# Distribution of win margins (runs)
data['win_by_runs'].plot(kind='hist',bins=20,color='orange',rwidth=0.8)
```
