```python
import pandas as pd
import numpy as np

# Aapka base data
raw_data = {
    'PassengerId': [1, 2, 3, 4, 5, 6, 7, 8],
    'Survived': [0, 1, 1, 0, 0, None, 1, 0],
    'Pclass': [3, 1, 3, 1, 3, 3, 1, 2],
    'Name': ['Braund', 'Cumings', 'Heikkinen', 'Futrelle', 'Allen', None, 'McCarthy', 'Palsson'],
    'Sex': ['male', 'female', 'female', 'female', 'male', 'male', None, 'male'],
    'Age': [22.0, 38.0, 26.0, 35.0, None, 54.0, 2.0, 27.0],
    'SibSp': [1, 1, 0, 1, 0, 0, 3, None],
    'Parch': [0, 0, 0, 0, 0, 0, 1, 1],
    'Ticket': ['A/5 21171', 'PC 17599', 'STON/O2', '113803', '373450', None, '17463', '349909'],
    'Fare': [7.25, 71.28, 7.92, 53.10, 8.05, 8.45, None, 21.07],
    'Cabin': [None, 'C85', None, 'C123', None, None, 'E46', None],
    'Embarked': ['S', 'C', 'S', 'S', 'S', 'Q', 'S', None]
}

data_repeated = {col: np.tile(val, 100) for col, val in raw_data.items()}
df = pd.DataFrame(data_repeated)

# --- Naya Data Add Karte Hain ---
# 1. Poora Khali Column
df['Remarks'] = np.nan 

# 2. Aik Poori Khali Row (Row number 800 par add karte hain)
df.loc[800] = [np.nan] * len(df.columns)

print(f"Data Shape: {df.shape}")
```

```python
df
```

```python
df_clean = df.dropna()
df_clean
```

```python
# 'how=all' ka matlab hai: "tabhi delete karo jab poori row khali ho"
df_clean = df.dropna(how='all')
df_clean
```

```python
df_clean.shape
```

```python
# axis=1 matlab Column. how='all' matlab jo poora khali ho.
df_no_empty_cols = df.dropna(axis=1, how='all')
df_no_empty_cols
```

```python
df_no_empty_cols.shape
```

```python
# Isse sirf wo rows jayengi jahan Name ya Age nahi likhi
df_final = df.dropna(subset=['Name', 'Age'])
df_final
```

```python
df_filled = df.fillna("Missing")
```

```python
df_filled
```

```python
df_cleaned = df.fillna(
    {
        "Age":0,
        "Sex":'Unknown',
        "Cabin":"No Cabin"
    }
)
```

```python
df_cleaned
```

```python
# Age ki khali jagah par sab logon ki average age bhar do
average_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(average_age)
```

```python
# Forward fill: Upar wali value niche copy hogi
df_filled = df.ffill()

# Check karne ke liye niche ye likhein
df_filled.head()
```
