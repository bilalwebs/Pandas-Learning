```python
import pandas as pd
```

```python
match = pd.read_csv("matches.csv")

match.head()
```

```python
delivery = pd.read_csv("deliveries.csv")
delivery.head()
```

```python
# Merge the two dataframes on the common column "id" and "match_id"

new = delivery.merge(match,left_on="match_id",right_on="id")
```

```python
print(delivery.shape)
print(match.shape)
```

```python
new
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset=['season'],keep='first').sort_values(by='season')[['season','batsman']]
```
